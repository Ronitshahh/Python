```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
import warnings 
warnings.filterwarnings("ignore")
```

### Lets First Load the dataset

```python
dataset = pd.read_excel("SuperStoreUS.xlsx")
```

### Display top 4 rows of the Dataset

```python
dataset.head(4)
```

### Find the Shape of the Dataset (No. of rows & columns)

```python
dataset.shape
```

### So now you know the number of orders

```python
print("number of rows",dataset.shape[0])
print("number of columns",dataset.shape[1])
```

### Now find the missing Values in the given dataset

```python
dataset.isnull().sum()
```

```python
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(),inplace=True)
```

### NOW AGAIN CHECK THE MISSING VALUES IN THE DATASET

```python
dataset.isnull().sum()
```

### SO NOW AS WE CAN SEE THERE IS NO NULL VALUES PRESENT IN THE DATASET

```python
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

### So as you can see there are 2 different objects (critical)

```python
dataset['Order Priority'] = dataset["Order Priority"].replace("Critical ","Critical")
```

```python
dataset['Order Priority'].value_counts()
```

### By replacing the object,you can see the problem is solved

```python
plt.figure(figsize=(5,4)) #to show the lenght and width 
sns.countplot(x="Order Priority",data=dataset)
plt.title("Count of Order Priority") #to show the title
plt.savefig("Count of Order Priority")
plt.show()
```

# So now you can see that with the help of seaborn library we created a countplot

# SHIP MODE

```python
dataset["Ship Mode"].value_counts()
```

```python
x = dataset["Ship Mode"].value_counts().index
y = dataset["Ship Mode"].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle= 60,autopct="%0.0f%%") # we can also change the decimal values by replacing 0
plt.legend(loc=2)
plt.show()
```

### As you can see the above classifications are also known as Univariate Analysis 
the other two are bivariate and multi-variate analysis

### Univariate analysis is a statistical method that examines each variable in a data set separately.

```python
### Now we are performing bivariate analysis with product category and shipping mode
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode",data=dataset,hue="Product Category")
plt.show()
```

### Customer Segment

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Segment",data=dataset)
plt.show()
```
